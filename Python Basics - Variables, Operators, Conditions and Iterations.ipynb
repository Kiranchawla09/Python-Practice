```python
# Find square root of a number without using math library
```

```python
def my_sqrt (start, end, num, precision):
    print (start, end, num, precision)
    guess = (start + end)/2
    print ("guess=", guess)
    error = abs((guess*guess) - num)
    if (error < precision):
        print ("stoping condition")
        return guess
    else:
        if ((guess*guess) > num): #over-estimation
            print ('over-estimation')
            return my_sqrt (start, guess, num, precision)
        else:
            print ('under-estimation')
            return my_sqrt (guess, end, num, precision)
```

```python
res = my_sqrt(0,5,5,0.01)
```
